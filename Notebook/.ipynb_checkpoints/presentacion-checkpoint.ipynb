### ¿Qué quiere saber ella?

1. ¿Cuándo y dónde podrá ver un eclipse?

```python
# Cargar el archivo de los trayectorias de eclipses
eclipses = gpd.read_file(os.path.join(data_pth, "Eclipses.shp"))
```

```python
# Verifique el sistema de referencia de coordenadas
eclipses.crs
```

```python
# Examina las primeras filas
eclipses.head()
```

```python
# Y vamos a trazar con un estilo personalizado
eclipses.plot(figsize=(15,10), color='black', edgecolor='yellow', alpha=0.75)
```

```python
# GeoPandas le permite trazar el cuadro de límite mínimo de cada entidad
eclipses.envelope.plot(figsize=(15,10))
```

```python
# Vamos a trazar los eclipses en la parte superior de nuestros datos mundiales
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
eclipses.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
base.set_axis_off()
```

```python
# Pero, ¿cómo sabemos cuál es cuál?
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))

# Podemos añadir un valor catórico y establecer la columna en Año. También podemos añadir una leyenda.
eclipses.plot(ax=base, cmap='tab10', alpha=0.5, categorical = True, column = "Year", legend=True)
base.set_axis_off()
base.get_legend().set_bbox_to_anchor((.05,.7))
```

### ¿Qué quiere saber ella?

1. ¿Cuándo y dónde podrá ver un eclipse?

2. ¿Qué próximo eclipse pasará por la mayoría de las personas?

```python
# Ahora trazaremos nuestro mapa base, nuestros trayectorias de eclipse y nuestras ciudades
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
eclipses.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
cities.plot(ax=base, color='orange', markersize=5)
base.set_axis_off()
```

```python
# Comience por obtener todas las ciudades que se intersecan con cualquier eclipse,
# una unión espacial utilizando la columna geometrica
allecities = gpd.sjoin(cities, eclipses, how='inner', op='intersects')
```

```python
allecities.head()
```

```python
allecities = pd.DataFrame(allecities[['POP', 'Year']])
```

```python
allecities.head()
```

```python
# Crear una variable temporal para mantener nuestros años y poblaciones
temp_group = allecities.groupby(['Year'])[["POP"]].sum()
```

```python
temp_group
```

```python
# Agregue esa columna a nuestros datos de eclipse, uniéndose a la columna Año
eclipses_pop = pd.merge(eclipses, temp_group, left_on='Year', right_index=True)
```

```python
# Ahora tenemos poblaciones asociadas a cada eclipse
eclipses_pop.head()
```

```python
# Ordena los resultados para saber qué eclipse cubrirá la mayoría de las personas
eclipses_pop.sort_values('POP', ascending = False)
```

```python
# Vamos a trazar esto
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
eclipses_pop.plot(ax=base, cmap='Oranges', alpha=0.5, column = "POP")
base.set_axis_off()
```

```python
# Como vimos anteriormente, este eclipse está ocurriendo en 2024
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
eclipses.plot(ax=base, cmap='tab10', alpha=0.5, categorical = True, column = "Year", legend=True)
base.set_axis_off()
base.get_legend().set_bbox_to_anchor((.05,.7))
```

### ¿Qué quiere saber ella?

1. ¿Cuándo y dónde podrá ver un eclipse?

2. ¿Qué próximo eclipse pasará por la mayoría de las personas?

3. ¿Cuáles son las ciudades más grandes en ese eclipse?

```python
# Asignemos el eclipse de 2024 a su propia variable
my_eclipse = eclipses[(eclipses['Year'] == 2024)]
```

```python
# Vamos a trazarlo 
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
my_eclipse.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
cities.plot(marker='*', color='orange', markersize=5, ax=base)
base.set_axis_off()
```

```python
# Trazar de nuevo, pero esta vez quiero acercarme a ese camino de eclipse
base = world.plot(color='lightgrey', linewidth=0.5, edgecolor='white', figsize=(15,10))
my_eclipse.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
cities.plot(marker='*', color='orange', markersize=5, ax=base)
bounds = my_eclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5])
base.set_axis_off()
```

```python
# Vamos a diseñar la trama para que el eclipse se vea mas "eclipse-y"
base = world.plot(color='lightgray', linewidth=0.5, edgecolor='white', figsize=(15,10))
my_eclipse.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
cities.plot(ax=base, color='black', markersize=20, edgecolor='white')
bounds = my_eclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5])
base.set_axis_off()
```

```python
# ¡Podemos ver que hay bastantes ciudades que se cruzan en mi camino! Pero queremos saber cuáles.
# Hacer una unión espacial para obtener la intersección
e_cities = gpd.sjoin(cities, my_eclipse, how='inner', op='intersects')
```

```python
# ¡Vamos a trazar los resultados!
base = world.plot(color='lightgray', linewidth=0.5, edgecolor='white', figsize=(15,10))
my_eclipse.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
e_cities.plot(ax=base, color='black', markersize=20, edgecolor='white')
bounds = my_eclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5])
base.set_axis_off()
```

```python
# También podemos dimensionar las ciudades en función de sus poblaciones.
base = world.plot(color='lightgray', linewidth=0.5, edgecolor='white', figsize=(15,10))
my_eclipse.plot(ax=base, color='black', edgecolor='yellow', alpha=0.75)
e_cities.plot(ax=base, color='black', edgecolor='white', markersize=cities['POP']/1000)
bounds = my_eclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5])
base.set_axis_off()
```

```python
print('¡Guauu! ¡Hay {} ciudades en el trayectoria del eclipse!'.format(len(e_cities)))
print("Aquí están los 5 más grandes:")
print(e_cities.sort_values('POP', ascending = False)['NAME'].head(5))
```

```python
top_five = e_cities.sort_values('POP', ascending = False)[:5]
```

```python
base = world.plot(color='lightgray', linewidth=0.5, edgecolor='white', figsize=(15,10))
my_eclipse.plot(ax=base, color='black', edgecolor='yellow', alpha=0.25)
top_five.plot(ax=base, color='black', edgecolor='white', markersize=cities['POP']/400)
for x, y, label in zip(top_five.geometry.x, top_five.geometry.y, top_five.NAME):
    base.annotate(label, xy=(x, y), xytext=(3, 3), textcoords="offset points")
bounds = my_eclipse.geometry.bounds
plt.xlim([bounds.minx.min()-5, bounds.maxx.max()+5])
plt.ylim([bounds.miny.min()-5, bounds.maxy.max()+5])
base.set_axis_off()
```

## [Folium](https://python-visualization.github.io/folium/)

#### Enlaza el poder de Python con leaflet.js

```python
import folium
```

```python
Cleveland = [41.4993, -81.6944]

osm_map = folium.Map(location=Cleveland, zoom_start=10)
osm_map
```

```python
map_stamen_toner = folium.Map(
    location=Cleveland,
    tiles='stamentoner',
    zoom_start=10
)
map_stamen_toner
```

```python
map_stamen_water = folium.Map(
    location=Cleveland,
    tiles='stamenwatercolor',
    zoom_start=10
)
map_stamen_water
```

```python
# Encontremos el centroide del eclipse que elegimos, para poder centrar nuestro mapa de folum en él.
x = my_eclipse.centroid.x.values[0]
y = my_eclipse.centroid.y.values[0]
print('y: ' + str(y) + ', x: ' + str(x))
```

```python
map_stamen_water = folium.Map(
    location=[y,x],
    tiles='stamenwatercolor',
    zoom_start=3
)

folium.GeoJson(top_five,name='Eclipse Cities').add_to(map_stamen_water)
folium.GeoJson(my_eclipse,name='Eclipse Path').add_to(map_stamen_water)

folium.LayerControl().add_to(map_stamen_water)
```

```python
map_stamen_water
```

```python
# Puedes escribir una función que crea un mapa con todas tus propiedades
# Éste crea popups para tus ciudades.

def plot_city_pop(top_five, my_eclipse):
    # generate a new map
    folium_map = folium.Map(location=[y, x], zoom_start=3, tiles="stamenwatercolor")
    folium.GeoJson(my_eclipse,name='Eclipse Path').add_to(folium_map)
    # for each row in the data, add a cicle marker
    for index, row in top_five.iterrows():
        
        # generate the popup message that is shown on click.
        popup_text = "Name: {}<br> Pop: {}"
        popup_text = popup_text.format(row["NAME"], row["POP"])
        
        folium.CircleMarker(location=
                            (row['geometry'].y, row['geometry'].x),
                            radius=6,popup=popup_text,fill=True).add_to(folium_map)        

    # Add a layer control if you like
    folium.LayerControl().add_to(folium_map)
    
    return folium_map
```

```python
# Llama a la función para crear el mapa
final_map = plot_city_pop(top_five, my_eclipse)
final_map
```

```python
# Save the results as a HTML file for easy sharing
final_map.save('mymap.html')
```
